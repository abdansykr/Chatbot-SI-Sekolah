```python
import pandas as pd
from langchain_community.document_loaders import DataFrameLoader

def process_csv(filepath, delimiter=","):
  """
  This function processes a single CSV file, converts it to a DataFrame,
  concatenates all columns into a single column with a delimiter,
  and returns the modified DataFrame with the combined data column.

  Args:
      filepath (str): The path to a single CSV file.
      delimiter (str, optional): The delimiter to use for separating concatenated columns (default: ",").

  Returns:
      pandas.DataFrame: The modified DataFrame with the combined data column.
  """
  try:
    # Read CSV file using pandas
    df = pd.read_csv(filepath)
    print(f"Successfully converted '{filepath}' to a pandas DataFrame.")

    # Get column names (optional, for reference)
    column_names = df.columns.tolist()

    # Concatenate all columns with the chosen delimiter
    df['combined_data'] = df.apply(lambda row: delimiter.join(row.astype(str)), axis=1)

    return df
  except Exception as e:
    print(f"Error processing file '{filepath}': {e}")
    return pd.DataFrame()  # Return empty DataFrame on error

def process_csv_files(csv_filepaths):
  """
  This function processes multiple CSV files, converts them to DataFrames,
  concatenates all columns in each DataFrame, and combines them into a single DataFrame.

  Args:
      csv_filepaths (list): A list containing paths to the CSV files.

  Returns:
      pandas.DataFrame: A single DataFrame containing combined data from all CSV files.
  """

  all_data = pd.DataFrame()  # Empty DataFrame to store combined data
  for filepath in csv_filepaths:
    try:
      # Process individual CSV file and get the modified DataFrame
      df = process_csv(filepath)
      all_data = pd.concat([all_data, df], ignore_index=True)  # Concatenate DataFrames
    except Exception as e:
      print(f"Error processing file '{filepath}': {e}")

  return all_data
```

```python
csv_files = ["/content/data_kurikulum.csv", "/content/jadwal_mata_pelajaran.csv", "/content/student_scores_expanded.csv"]
all_documents = process_csv_files(csv_files)
```

```python
loader = DataFrameLoader(all_documents, page_content_column="combined_data")
docs = loader.load()
```

```python
from langchain_community.vectorstores import Qdrant
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

url = os.getenv("QDRANT_URL")
api_key = os.getenv("QDRANT_KEY")

qdrant = Qdrant.from_documents(
    documents=docs,
    embedding=embeddings,
    url=url,
    collection_name="chatbot",
    api_key=api_key
)
```

```python
query = "What is so special about Mistral 7B?"
qdrant.similarity_search(query, k=3)
```

```python
def custom_prompt(query: str):
    results = qdrant.similarity_search(query, k=3)
    source_knowledge = "\n".join([x.page_content for x in results])
    augment_prompt = f"""Using the contexts below, answer the query:

    Contexts:
    {source_knowledge}

    Query: {query}"""
    return augment_prompt

print(custom_prompt(query))
```

```python
prompt = HumanMessage(
    content=custom_prompt(query)
)

messages.append(prompt)

res = chat.invoke(messages)

print(res.content)
```
